```python
import numpy as np
import pylab as plt
```

```python
#Loading in data from file Efield_generator.ipynb

X = np.loadtxt("X(5000).txt") #load in the DATA 
labels = np.loadtxt("labels(5000).txt") #load in the number of charges 
```

```python
#This plots the first image inside the data set
plt.imshow(X[0].reshape(20,20))
plt.show()
```

```python
Cx = np.mean(X[:])
for i in range(400):
    m = np.mean(X[i])
    X[i] = X[i] - Cx
```

```python
Cov = np.matmul(X.T,X)
vals, vecs = np.linalg.eig(Cov)
plt.plot(vals,'o')
```

```python
P1 = np.zeros(len(X[:]))
P2 = np.zeros(len(X[:]))
for i in range(len(X[:])):
    P1[i] = np.dot(X[i],vecs[:,0])
    P2[i] = np.dot(X[i],vecs[:,1])
```

```python
plt.figure(figsize=(7,7))
plt.scatter(P1,P2,c = labels, marker = 'o', linewidth= 5,alpha = 0.5, s= 5)
h = plt.colorbar()
h.set_label('N-Charges', y=.5, rotation=90, size = 25)
plt.xlabel("$P_1$", size =25)
plt.ylabel("$P_2$", size =25)
plt.show()
```
